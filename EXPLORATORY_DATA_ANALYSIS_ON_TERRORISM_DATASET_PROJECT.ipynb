```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# loading data
```

```python
data=pd.read_csv("C:\\Users\\ADMIN\\Downloads\\globalterrorismdb_0718dist.csv",low_memory=False,)
data.head(7)
```

```python
data.shape
```

```python
data.describe()
```

```python
# features selectio based on required exploratory analysis
```

```python
df=data[['iyear','imonth','iday','country_txt','region_txt','provstate','city','latitude','longitude','location','attacktype1_txt','targtype1_txt','weaptype1_txt','gname','nkill','nwound','motive','addnotes']]
```

```python
df.rename(columns={'iyear':'year','imonth':'month','iday':'day','country_txt':'Country','region_txt':'Region','provstate':'Province/State','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','gname':'Group_name','nkill':'killed','nwound':'wounded'},inplace=True)
```

```python
df.head()
```

```python
df['wounded']=df['wounded'].fillna(0)
df['killed']=df['killed'].fillna(0)
df['Casualties']=df['killed']+df['wounded']
df.head()
```

```python
df['Country'].value_counts()[:10]
```

```python
plt.subplots(figsize=(10,5))
sns.barplot(df['Country'].value_counts()[:10].index,df['Country'].value_counts()[:10].values,palette="hls")
plt.title("Top Countries Affected")
plt.xlabel("Countries")
plt.ylabel("Count")
plt.xticks(rotation=50)
plt.show()
```

```python
df['Group_name'].value_counts()[:10]
```

```python
df.city.value_counts().to_frame().sort_values('city',axis=0,ascending=False).drop("Unknown").head(10).plot(kind='bar',figsize=(10,5),color='green')
plt.xticks(rotation=50)
plt.xlabel("City")
plt.ylabel("Number of attacks")
plt.title("Top 10 most affected cities",fontsize=10)
plt.show()
```

```python
datafilter=df.groupby(['Group_name','Country']).sum().sort_values('killed',ascending=True).drop('Unknown').reset_index().tail()
datafilter[['Group_name','Country','killed']]
```

```python
plt.subplots(figsize=(10,7))
sns.countplot(df.Target_type,order=df["Target_type"].value_counts().index);
plt.title("Most attacked target type")
plt.xticks(rotation=90)
plt.xlabel("Target type",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.show()
```

```python
df["Attack_type"].value_counts().to_frame().sort_values('Attack_type',ascending=False).plot(kind="bar",color="blue",figsize=(10,5))
plt.xticks(rotation=75)
plt.xlabel("Attack types",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Most used attack types",fontsize=20)
plt.show()
```

```python
df['Province/State'].isna().sum()
```

```python
dd=df[(df["Province/State"]!='Unknown')]
ds=dd["Province/State"].dropna(axis=0).value_counts()[:15]
plt.subplots(figsize=(15,7))
g=sns.barplot(ds.index,ds.values,palette="crest")
for val in g.patches:
    g.annotate(format(val.get_height()),(val.get_x() + val.get_width()/2, val.get_height()),
               ha='center',size=10,va='center',textcoords='offset points',xytext=(0, 8))
plt.title("Most attacked Province/State",fontsize=20)
plt.xlabel("Province/State",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.xticks(rotation=90,fontsize=10)
plt.show()
```

```python
df["Region"].value_counts().to_frame().sort_values('Region',ascending=False).plot(kind="bar",color='grey',figsize=(10,5))
plt.xticks(rotation=75)
plt.xlabel("Regions",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Most attacked regions",fontsize=20)
plt.show()
```

```python
df[['Group_name','killed']].groupby(['Group_name'],axis=0).sum().drop('Unknown').sort_values('killed',ascending=False)[:10].plot(kind='bar',color="red",figsize=(15,7))
plt.title("Top 10 terrorist groups",fontsize=15)
plt.xlabel("Terrorist attack groups",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.xticks(rotation=75)
plt.show()
```
